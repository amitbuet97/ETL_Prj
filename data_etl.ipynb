```python
# importing dependencies

import pandas as pd

import numpy as np

from sqlalchemy import create_engine
```

### Store CSV into DataFrame

```python
#importing 2 csv files from IMDB and Rotten Movies 

csv_file = "Resources/movies_imbd.csv"

csv_file2 = "Resources/movies_rotten.csv"

imdb_data_df = pd.read_csv(csv_file)

imdb_data_df.head()
```

```python
# Select Columns for the datatype 

movie_table_df = imdb_data_df[["movie_title", "director_name","imdb_score", "actor_1_name","actor_2_name","actor_3_name","gross","budget"
                           ,"movie_facebook_likes","country" ]]

#Rename the column name

movie_table_df = movie_table_df.rename(columns = {"movie_title":"Movie_Title","director_name":"Director_Name",
                                       "imdb_score":"IMDB_score","actor_1_name":"Actor_1","actor_2_name":"Actor_2",
                                        "actor_3_name":"Actor_3","gross":"Gross_Revenue","budget":"Budget",
                                        "movie_facebook_likes":"Total_Facebook_Likes","country":"Country"})
# Check thb dataframe

movie_table_df.head()
```

```python
# remove the rows with blanc values

movie_table_df.dropna(inplace=True)
movie_table_df.head()
```

```python
# read the rotten movies csv file

rot_data_df = pd.read_csv(csv_file2)

rot_data_df.head()
```

```python
# List all the columns in the table
rot_data_df.columns
```

```python
# Select Columns for the datatype 

rot_data_df = rot_data_df[["Title", "MPAA Rating","Distributor", "Major Genre","Rotten Tomatoes Rating" ]]

# rename the column

rot_data_df = rot_data_df.rename(columns = {"Title":"Movie_Title","Major Genre":"Genre",
                                       "Rotten Tomatoes Rating":"Rotten_Tomatoes_Rating","MPAA Rating":"MPAA"})

rot_data_df.head()
```

```python
#replace NAN with 0.

rot_data_df.fillna(0,inplace=True)
```

```python
rot_data_df.dtypes
```

```python
# Export CSV


rot_data_df.to_csv('Rotten_Tomatoes.csv',index=False)

movie_table_df.to_csv('imdb.csv',index=False)


```
